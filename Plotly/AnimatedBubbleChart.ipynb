```python
import io
import pandas as pd
import numpy as np

from IPython.display import display
import plotly.offline as py
import plotly.express as px
import plotly.graph_objs as go
import plotly.tools as tls
from plotly.offline import init_notebook_mode, iplot
py.init_notebook_mode(connected=True)

import warnings
warnings.filterwarnings('ignore')

from __future__ import division
```

```python
def configure_plotly_browser_state():
  import IPython
  display(IPython.core.display.HTML('''
        <script src="/static/components/requirejs/require.js"></script>
        <script>
          requirejs.config({
            paths: {
              base: '/static/base',
              plotly: 'https://cdn.plot.ly/plotly-latest.min.js?noext',
            },
          });
        </script>
        '''))
```

```python
gapminder_indicators = pd.read_csv(io.BytesIO(gapminder_uploaded['gapminder.tsv']), delimiter='\t')
honey_production = pd.read_csv(io.BytesIO(honey_uploaded['honeyproduction.csv']))
world_happiness = pd.read_csv(io.BytesIO(happiness_uploaded['2017.csv']))
```

```python
configure_plotly_browser_state()

fig = px.scatter(
    gapminder_indicators,
    x = 'gdpPercap',
    y = 'lifeExp',
    animation_frame = 'year',
    animation_group = 'country',
    size = 'pop',
    color = 'continent',
    title = 'Gapminder GLobal Indicators',
    log_x = True,
    size_max = 55,
    range_x = [100, 100000],
    range_y = [25, 90],
    hover_name = 'country',
    labels = dict(
      lifeExp = 'Life Expectancy',
      gdpPercap = 'GDP per Capita'
    ),
  )

fig.update_layout(template='plotly_white')

iplot(fig, config={'scrollzoom': True})
```

```python
configure_plotly_browser_state()

fig = px.scatter(
    honey_production,
    x = 'totalprod',
    y = 'prodvalue',
    animation_frame = 'year',
    animation_group = 'state',
    size = 'yieldpercol',
    color = 'numcol',
    title = 'Timeline of honey production for each state in USA',
    log_x = True,
    log_y = True,
    labels = dict(
      totalprod = 'Total Production',
      prodvalue = 'Production Value',
      numcol = 'Colonies',
    ),
    hover_name = 'state',
  )

fig.update_layout(template='plotly_white')

iplot(fig, config={'scrollzoom': True})
```

```python
configure_plotly_browser_state()

fig = px.scatter(
    honey_production,
    x = 'totalprod',
    y = 'priceperlb',
    animation_frame = 'year',
    animation_group = 'state',
    size = 'yieldpercol',
    color = 'numcol',
    title = 'Timeline of honey production and its prices for each state in USA',
    log_x = True,
    log_y = True,
    labels = dict(
      totalprod = 'Total Production',
      priceperlb = 'Price per pound',
      numcol = 'Colonies',
    ),
    color_continuous_scale = 'Sunset',
    hover_name = 'state',
  )

fig.update_layout(template='plotly_white')

iplot(fig, config={'scrollzoom': True})
```
