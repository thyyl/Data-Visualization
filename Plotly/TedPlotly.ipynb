Plotly Code Loading

```python
def configure_plotly_browser_state():
  import IPython
  display(IPython.core.display.HTML('''
        <script src="/static/components/requirejs/require.js"></script>
        <script>
          requirejs.config({
            paths: {
              base: '/static/base',
              plotly: 'https://cdn.plot.ly/plotly-latest.min.js?noext',
            },
          });
        </script>
        '''))
```

1. Line Chart

```python
def LineChart(df):
  configure_plotly_browser_state()
  trace = go.Scatter(
      x = df.index,
      y = df.comments,
      mode = 'lines',
      name = 'comments',
      marker = dict(color='rgba(80, 26, 80, 0.8)'),
      text = df.main_speaker,
  )

  data = [trace]

  layout = go.Layout(
      title = 'Comments for Ted Talks',
      xaxis = dict(title='Index', ticklen=5, zeroline=False),
      template = 'plotly_white',
  )

  fig = go.Figure(data=data, layout=layout)
  iplot(fig, filename='Line Chart')
```

```python
LineChart(df)
```

2. Scatter Plot

```python
def ScatterPlot(dfEvents):
  configure_plotly_browser_state()
  data = []

  for name, df, color in dfEvents:
    trace = go.Scatter(
        x = df.index,
        y = df.views,
        mode = 'markers',
        name = name,
        marker = dict(color=color, size=10),
        text = df.main_speaker,
    )

    data.append(trace)

  layout = go.Layout(
      title = 'Views of TEDGlobal 2005, TED2002, and Royal Institution',
      xaxis = dict(title='Index', ticklen=5, zeroline=False),
      yaxis = dict(title='Views', ticklen=5, zeroline=False),
      template = 'plotly_white',
  )

  fig = go.Figure(data=data, layout=layout)
  fig.show()
```

```python
dfGlobal = df[df['event'] == 'TEDGlobal 2005']
colorGlobal = 'rgba(255, 128, 255, 0.8)'
df2002 = df[df['event'] == 'TED2002']
color2002 = 'rgba(255, 128, 2, 0.8)'
dfRoyal = df[df['event'] == 'Royal Institution']
colorRoyal = 'rgba(0, 255, 200, 0.8)'

dfEvents = [('TEDGlobal 2005', dfGlobal, colorGlobal), ('TED2002', df2002, color2002), ('Royal Institution', dfRoyal, colorRoyal)]
ScatterPlot(dfEvents)
```

3. Bubble Chart

```python
def BubbleChart(df):
  configure_plotly_browser_state()
  df.index = range(0, len(df))

  trace = go.Scatter(
      x = df.index,
      y = df.views,
      mode = 'markers',
      marker = dict(color=df.duration, size=df.comments, showscale=True),
      text = df.main_speaker,
  )

  data = [trace]

  layout = go.Layout(template = 'plotly_white')

  fig = go.Figure(data=data, layout=layout)
  fig.show()
```

```python
df_sorted = df.sort_values(by='views', ascending=True)
df_sorted = df_sorted.iloc[:20, :]

BubbleChart(df_sorted)
```

4. Insets Plot

```python
import datetime
```

```python
def InsetsPlot(df):
  configure_plotly_browser_state()
  data_duration2 = []
  data_duration2 = [str(datetime.timedelta(seconds=i)) + ' minute ' for i in df.duration]
  df_100 = df.iloc[:100, :]

  trace1 = go.Scatter(
      x = df_100.index,
      y = df_100.views,
      name = 'Views',
      marker = dict(color = 'rgba(200, 75, 45, 0.8)'),
  )

  trace2 = go.Scatter(
      x = df_100.index,
      y = df_100.duration,
      xaxis = 'x2',
      yaxis = 'y2',
      name = "duration",
      text = data_duration2,
      marker = dict(color = 'rgba(85, 20, 200, 0.8)'),
  )
  
  data = [trace1, trace2]
  layout = go.Layout(
      xaxis2=dict(
          domain=[0.6, 0.95],
          anchor='y2',        
      ),
      yaxis2=dict(
          domain=[0.6, 0.95],
          anchor='x2',
      ),
      title = 'Views and Comments',
      template = 'plotly_white',
  )
  
  fig = go.Figure(data=data, layout=layout)
  plt.savefig('graph.png')
  iplot(fig)
  plt.show()
```

```python
InsetsPlot(df)
```
