1. Plotly Distribution Plot

Distribution plot of one stats 

```python
def PlotlyDistPlot(df, nameOfFeature):
  trace = go.Histogram(x = df[nameOfFeature])

  layout = go.Layout(
    title = '{} Distribution Plot'.format(nameOfFeature),
    template = 'plotly_white'
  )

  data = [trace]

  figure = go.Figure(data=data, layout=layout)
  figure.show()
```

2. Boxplot

```python
def BoxPlot(df, firstFeature, secondFeature, thirdFeature, fourthFeature, fifthFeature, sixthFeature):
  configure_plotly_browser_state()

  trace0 = go.Box(y=df[firstFeature] , name=firstFeature)
  trace1 = go.Box(y=df[secondFeature], name=secondFeature)
  trace2 = go.Box(y=df[thirdFeature] , name=thirdFeature)
  trace3 = go.Box(y=df[fourthFeature], name=fourthFeature)
  trace4 = go.Box(y=df[fifthFeature] , name=fifthFeature)
  trace5 = go.Box(y=df[sixthFeature] , name=sixthFeature)

  data = [trace0, trace1, trace2,trace3, trace4, trace5]

  layout = go.Layout(
    title = 'ALl Stats BoxPlot',
    template = 'plotly_white'
  )

  figure = go.Figure(data=data, layout=layout)
  figure.show()
```

```python
BoxPlot(df, 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed')
```

Self-customized boxplot

```python
def SelfCustomizedBoxplot(df, firstFeature, secondFeature, thirdFeature, fourthFeature, fifthFeature, sixthFeature):
  configure_plotly_browser_state()
  
  trace0 = go.Box(
      y = df[firstFeature], 
      name = firstFeature + '(Mean)', 
      boxmean = True,
  )
  
  trace1 = go.Box(
      y=df[secondFeature], 
      name=secondFeature + '(Mean and SD)',
      boxmean='sd',
  )
  
  trace2 = go.Box(
      y = df[thirdFeature], 
      name = thirdFeature + '(All Points)',
      boxpoints = 'all',
      pointpos = -2,
      jitter = 0.5,
  )
 
  trace3 = go.Box(
      y = df[fourthFeature],
      name = fourthFeature + '(Only Whiskers)',
      boxpoints = False,
  )
  
  trace4 = go.Box(
      y = df[fifthFeature], 
      name = fifthFeature + '(Suspected Outliers)',
      boxpoints = 'suspectedoutliers',
      marker = dict(
        outliercolor = 'rgba(219, 64, 82, 0.6)',
        line = dict(
            outliercolor = 'rgba(219, 64, 82, 0.6)',
            outlierwidth = 2)
      ),
      line = dict(color = 'rgb(8,81,156)'),
  )
  
  trace5 = go.Box(
      y = df[sixthFeature], 
      name = sixthFeature + '(Whiskers and Outliers)',
      boxpoints = 'outliers',
      line = dict(color = 'rgb(107,174,214)'),
  )

  data = [trace0, trace1, trace2,trace3, trace4, trace5]

  layout = go.Layout(
    title = 'ALl Stats BoxPlot',
    template = 'plotly_white',
  )

  figure = go.Figure(data=data, layout=layout)
  figure.show()
```

```python
SelfCustomizedBoxplot(df, 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed')
```

3. Radar Charts

```python
def RadarChartSpecific(df, pokemon):
  configure_plotly_browser_state()
  x = df[df['Name'] == pokemon]

  trace = go.Scatterpolar(
      r = [x['HP'].values[0], x['Attack'].values[0], x['Defense'].values[0], x['Sp. Atk'].values[0], x['Sp. Def'].values[0], x['Speed'].values[0]],
      theta = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'HP'],
      fill = 'toself',
  )

  layout = go.Layout(
      polar = dict(
          radialaxis = dict(visible=True, range=[0, 250])
      ),
      showlegend = False,
      title = 'Stats of {}'.format(pokemon),
      template = 'plotly_white',
  )

  data = [trace]
  figure = go.Figure(data=data, layout=layout)
  iplot(figure, filename='Single Pokemon Stats')
```

```python
RadarChartSpecific(df, 'Mewtwo')
```

Comparing between two pokemons

```python
def RadarChartCompare(df, firstPokemon, secondPokemon):
  configure_plotly_browser_state()
  x = df[df['Name'] == firstPokemon]
  y = df[df['Name'] == secondPokemon]

  trace0 = go.Scatterpolar(
      r = [x['HP'].values[0], x['Attack'].values[0], x['Defense'].values[0], x['Sp. Atk'].values[0], x['Sp. Def'].values[0], x['Speed'].values[0]],
      theta = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'HP'],
      fill = 'toself',
      name = firstPokemon
  )

  trace1 = go.Scatterpolar(
      r = [y['HP'].values[0], y['Attack'].values[0], y['Defense'].values[0], y['Sp. Atk'].values[0], y['Sp. Def'].values[0], y['Speed'].values[0]],
      theta = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'HP'],
      fill = 'toself',
      name = secondPokemon
  )

  layout = go.Layout(
      polar = dict(
          radialaxis = dict(visible=True, range=[0, 250])
      ),
      showlegend = True,
      title = 'Stats of {} vs {}'.format(firstPokemon, secondPokemon),
      template = 'plotly_white',
  )

  data = [trace0, trace1]
  figure = go.Figure(data=data, layout=layout)
  iplot(figure, filename='Single Pokemon Stats')
```

```python
RadarChartCompare(df, 'Dialga', 'Palkia')
```

4. Scatterplot

```python
def ScatterPlotStats(df, firstStats, secondStats, thirdStats):
  configure_plotly_browser_state()
  trace = go.Scatter(
      x = df[firstStats],
      y = df[secondStats],
      mode = 'markers',
      marker = dict(
          color = df[thirdStats],
          colorscale = 'emrld',
          showscale = True,
      ),
      text = df['Name'],
  )

  data = [trace]

  layout = go.Layout(
      showlegend = False,
      font = dict(family='Courier New, monospace', size=10, color='black'),
      title = 'Scatter Plot of {} vs {} with {} as colourscale'.format(firstStats, secondStats, thirdStats),
      template = 'plotly_white',
  )

  fig = go.Figure(data=data, layout=layout)
  iplot(fig, filename = 'Scatterplot')
```

```python
ScatterPlotStats(df, 'Attack', 'Defense', 'Speed')
```

3D ScatterPlot

```python
def ScatterPlot3D(df, firstStats, secondStats, thirdStats):
  configure_plotly_browser_state()
  trace = go.Scatter3d(
    x = df[firstStats],
    y = df[secondStats],
    z = df[thirdStats],
    mode = 'markers',
    marker = dict(
        size = 4,
        line = dict(
            color = 'rgb(60, 179, 113)',
            width = 0.5,
        ),
        color = 'rgb(60, 179, 113)',
    ),
    opacity = 1,
  )

  data = [trace]

  layout = go.Layout(
      margin = dict(l=0, r=0, t=0, b=0),
      xaxis = dict(title=firstStats),
      yaxis = dict(title=secondStats),
      title = '{} vs {} vs {}'.format(firstStats, secondStats, thirdStats),
      template = 'plotly_white',
  )

  fig = go.Figure(data=data, layout=layout)
  iplot(fig, filename = '3D Scatterplot')
```

```python
ScatterPlot3D(df, 'Attack', 'Defense', 'Speed')
```

5. Bubble Charts

```python
def BubbleCharts(df, types, firstStats, secondStats):
  configure_plotly_browser_state()
  sizeref = 2 * max(df['HP']) / 3000
  data = []

  for i in range(1, 6):
    if i == 1:
      generation = 'First'
    elif i == 2:
      generation = 'Second'
    elif i == 2:
      generation = 'Third'
    elif i == 2:
      generation = 'Fourth'
    elif i == 2:
      generation = 'Fifth'
    else:
      generation = 'Sixth'

    trace = go.Scatter(
        x = df[firstStats][df['Type 1'] == types][df['Generation'] == i],
        y = df[secondStats][df['Type 1'] == types][df['Generation'] == i],
        mode = 'markers',
        name = generation + 'Generation',
        text = df['Name'][df['Type 1'] == types][df['Generation'] == i],
        marker = dict(
            symbol = 'circle',
            sizemode = 'area',
            size = df['HP'][df['Type 1'] == types][df['Generation'] == i],
            sizeref = sizeref,
            line = dict(width=2),
        )
    )
    
    data.append(trace)
    
  layout = go.Layout(
      xaxis = dict(title=firstStats),
      yaxis = dict(title=secondStats),
      title = '{} vs {} of {} Pokemons Over Generations'.format(firstStats, secondStats, types),
      template = 'plotly_white',
  )

  fig = go.Figure(data=data, layout=layout)
  iplot(fig, filename = 'Bubble Chart')
```

```python
BubbleCharts(df, 'Fire', 'Attack', 'Defense')
```

6. Violin Plots

```python
def ViolinPlots(df):
  configure_plotly_browser_state()
  data = []

  for i in range(5, 11):
    trace = {
        'type': 'violin',
        'x': max(df.iloc[:, i]),
        'y': df.iloc[:, i],
        'name': list(df.columns)[i],
        'box': {'visible': True},
        'meanline': {'visible': True},
    }

    data.append(trace)

    fig = {
      'data': data,
      'layout' : {
          'title': "Violin plot of all stats",
          'yaxis': {
              'zeroline': False,
          }
      }
    }

  iplot(fig, filename = 'Violin Plot', validate=False)
```

```python
ViolinPlots(df)
```

```python
def TreeMap(df):
  configure_plotly_browser_state()
  x = 0
  y = 0
  width = 50
  height = 50
  type_list = list(df['Type 1'].unique())
  values = [len(df[df['Type 1'] == i]) for i in type_list]

  normed = squarify.normalize_sizes(values, width, height)
  rectangles = squarify.squarify(normed, x, y, width, height)

  color_brewer = ['#2D3142','#4F5D75','#BFC0C0','#F2D7EE','#EF8354','#839788','#EEE0CB','#BAA898','#BFD7EA','#685044','#E9AFA3','#99B2DD','#F9DEC9','#3A405A','#494949','#FF5D73','#7C7A7A','#CF5C36','#EFC88B']
  shapes = []
  annotations = []
  counter = 0

  for rectangle in rectangles:
    shapes.append(
        dict(
          type = 'rect',
          x0 = rectangle['x'],
          y0 = rectangle['y'],
          x1 = rectangle['x'] + rectangle['dx'],
          y1 = rectangle['y'] + rectangle['dy'],
          line = dict(width=2),
          fillcolor = color_brewer[counter] 
        )
    )

    annotations.append(
        dict(
            x = rectangle['x'] + (rectangle['dx']/2),
            y = rectangle['y'] + (rectangle['dy']/2),
            text = '{} - {}'.format(type_list[counter], values[counter]),
            showarrow = False
        )
    )

    counter = counter + 1
    if counter >= len(color_brewer):
      counter = 0

  trace = go.Scatter(
      x = [rectangle['x'] + (rectangle['dx'] / 2) for rectangle in rectangles],
      y = [rectangle['y'] + (rectangle['dy'] / 2) for rectangle in rectangles],
      text = [str(value) for value in values],
      mode = 'text',
  )

  layout = dict(
      height = 700,
      width = 700,
      xaxis=dict(showgrid=False, zeroline=False),
      yaxis=dict(showgrid=False, zeroline=False),
      shapes=shapes,
      annotations=annotations,
      hovermode='closest',
      font=dict(color="#FFFFFF"),
      template = 'plotly_white'
  )

  figure = dict(data=[trace], layout=layout)
  iplot(figure, filename='Squarify Treemap')
```

```python
TreeMap(df)
```

7. Bullet Chart

```python
def BulletChart(df, pokemon):
  configure_plotly_browser_state()
  x = df[df['Name'] == pokemon]
  data = []
  stats = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

  for stat in stats:
    trace = ({
        'label': stat, 
        'sublabel': x[stat].values[0],
        'range': [int(max(df[stat]) * 0.5), int(max(df[stat]) * 0.75), int(max(df[stat]))],
        'performance': [int(max(df[stat]) * 0.55), int(max(df[stat]) * 0.7)],
        'point': [x[stat].values[0]],
    })

    data.append(trace)

  figure = ff.create_bullet(
      data,
      titles = 'label',
      subtitles = 'sublabel',
      markers = 'point',
      measures = 'performance',
      ranges = 'range',
      orientation = 'v',
      width = 800,
      height = 800,
  )

  iplot(figure, filename='Bullet Chart')
```

```python
BulletChart(df, 'Blaziken')
```
